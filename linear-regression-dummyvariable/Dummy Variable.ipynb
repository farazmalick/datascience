```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sb
sb.set()
```

```python
raw_data=pd.read_csv("1.03. Dummies.csv")
raw_data
```

```python
data=raw_data.copy()
```

```python
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
y=data['GPA']
x1=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
newdata=pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
newdata=newdata[['const','SAT','Attendance']]
newdata
```

```python
prediction=result.predict(newdata)
prediction
```

```python
pred=pd.DataFrame({'prediction':prediction})
fdata=newdata.join(pred)
fdata
```

```python
plt.scatter(data['SAT'],y)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
yhat=0.0017*data['SAT']+0.275
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='green')
fig=plt.plot(data['SAT'],yhat,lw=2,c='yellow')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
